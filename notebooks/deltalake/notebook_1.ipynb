# **delta table creation**

using sql api

```python
%sql
CREATE TABLE deltalakenishant.default.first_table(
  id INT,
  salary INT
)
```

## using delta api

```python
from delta.tables import DeltaTable
```

```python
DeltaTable.createIfNotExists(spark) \
    .tableName("deltalakenishant.default.firstdeltaapi") \
    .addColumn("id", "INT") \
    .addColumn("salary", "INT") \
    .execute()
```

```python
%sql
DROP TABLE deltalakenishant.default.firstdeltaapi
```

# **generated **columns****

### # identity columns

```python
from delta.tables import DeltaTable, IdentityGenerator
from pyspark.sql.types import LongType, IntegerType, StringType

DeltaTable.create(spark).tableName("deltalakenishant.default.firstdeltaapi") \
 .addColumn("id_col1", dataType=LongType(), generatedAlwaysAs=IdentityGenerator()) \
 .addColumn("salary", dataType=IntegerType()) \
 .addColumn("name", dataType=StringType()) \
 .execute()
```

computed columns

```python
DeltaTable.create(spark).tableName("deltalakenishant.default.seconddeltaapi") \
 .addColumn("salary", dataType=IntegerType()) \
 .addColumn("name", dataType=StringType()) \
 .addColumn("salaryaftertax", dataType=LongType(), generatedAlwaysAs="CAST((salary*0.8) AS BIGINT)") \
 .execute()
```

```python
%sql
Insert into deltalakenishant.default.seconddeltaapi(salary,name)
values
(1000,'a'),
(1100,'b')
```
