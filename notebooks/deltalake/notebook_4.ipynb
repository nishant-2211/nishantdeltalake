# upsert

```python
data = [(1,100,'xyz',1),(9,200,'bb',1),(10,300,'cc',1)]
df = spark.createDataFrame(data,['cus_id','income','name','tip'])

display(df)
```

```python
from delta.tables import DeltaTable
```

```python
dlt_obj =DeltaTable.forPath(spark,"/Volumes/deltalakenishant/default/deltavol/dmlsink/")

dlt_obj.alias("trg").merge(df.alias("src"),"trg.cus_id = src.cus_id")\
      .whenMatchedUpdateAll()\
      .whenNotMatchedInsertAll()\
      .execute()
```
