```python
data = [(1,100,"aa",1),(2,200,"bb",1),(3,300,"cc",1),(4,400,"dd",1)]

df = spark.createDataFrame(data,['id','salary','name', 'sus_column'])
                                   
display(df)
                                 
```

```python
df.write.format("delta") \
        .mode("append")\
        .save("/Volumes/deltalakenishant/default/deltavol/demosink/")
```

```python
df.write.format("delta") \
        .mode("append")\
        .option("mergeSchema", True) \
        .save("/Volumes/deltalakenishant/default/deltavol/demosink/")
```
